```python
import pandas as pd

df = pd.read_csv('scopus_covid.csv')
display(df.head())
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Year'], bins=range(int(df['Year'].min()), int(df['Year'].max()) + 2), kde=False)
plt.title('Número de Publicaciones por Año')
plt.xlabel('Año')
plt.ylabel('Número de Publicaciones')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each 'Source title'
source_counts = df['Source title'].value_counts().head(10) # Display top 10 for clarity

plt.figure(figsize=(12, 7))
sns.barplot(x=source_counts.values, y=source_counts.index, palette='viridis')
plt.title('Top 10 Revistas con Mayor Número de Publicaciones')
plt.xlabel('Número de Publicaciones')
plt.ylabel('Revista')
plt.tight_layout()
plt.show()
```
